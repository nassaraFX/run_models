### **Situation**

You own a small company.

You expect profits

$
\begin{array}{|c|c|}
\hline
\textbf{Year} & \textbf{Profit (\$)} \\ \hline
1 & 100 \\ \hline
2 & 120 \\ \hline
3 & 150 \\ \hline
4 & 180 \\ \hline
5 & 200 \\ \hline
\end{array}
$

Discount rate (required return) = 10%

- **Formula**

Present Value (PV)
$$PV = \frac{\text{Profit}}{(1 + r)^t}$$

$
\text{Where:}
$
$
r = 0.10
$
$
t = \text{year}
$

- **Calculation**

$$
\begin{array}{|c|c|c|c|}
\hline
\textbf{Year} & \textbf{Profit} & \textbf{PV Calculation} & \textbf{PV} \\ \hline
1 & 100 & 100 / (1.1)^1 & 90.91 \\ \hline
2 & 120 & 120 / (1.1)^2 & 99.17 \\ \hline
3 & 150 & 150 / (1.1)^3 & 112.70 \\ \hline
4 & 180 & 180 / (1.1)^4 & 122.94 \\ \hline
5 & 200 & 200 / (1.1)^5 & 124.18 \\ \hline
\end{array}
$$

- **Total Firm Value**
$$
\begin{aligned}
\text{Firm Value} &= \sum PV \\
&= 90.91 + 99.17 + 112.70 + 122.94 + 124.18 \\
&= 549.90
\end{aligned}
$$

- **Interpretation**

***Future profits*** = 100, 120, 150, 180, 200

***Today value*** = 549.90

So, ***the company is worth about $549.90 today***

- **Calculate PV**
- **Plot profits vs present values**

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
years = np.array([1, 2, 3, 4, 5])
profits = np.array([100, 120, 150, 180, 200])
r = 0.10

# Present Value calculation
pv = profits / (1 + r) ** years

# Total value
total_value = pv.sum()
print("Total Firm Value:", total_value)

# Plot
plt.figure()
plt.plot(years, profits, marker='o', label="Future Profits")
plt.plot(years, pv, marker='o', linestyle='--', label="Present Value")

plt.xlabel("Years")
plt.ylabel("Value ($)")
plt.title("Future Profits vs Present Value")
plt.legend()
plt.grid()

plt.show()
```

**The most important observation**

As years increase:

Blue line goes up (profits grow)

Orange line grows slowly

Why?

Because of discounting

Look at Year 5

Future profit = 200

Present value ≈ 124

This means:

Even if you will receive 200, today it is only worth 124

**Why the gap increases**

The distance between the two lines becomes bigger because

Time increases
Risk and waiting increase
So value today decreases more
Final meaning of the whole graph

*This graph is telling*

A firm’s value is NOT the sum of future profits
It is the sum of discounted profits

***One sentence***

The value of a firm is equal to the present value of all expected future profits because money received in the future is worth less than money received today.

### **Theory of Consumer Choice**

1. **Budget Constraint**

The budget constraint represents all combinations of goods a consumer can afford given their income and prices.

It is written as:

$$P_x X + P_y Y = I$$

$\text{Where:}$
$$X, Y = \text{quantities of goods}$$
$$P_x, P_y = \text{prices}$$
$$I = \text{income}$$

- Usefull for Graph

$$Y = \frac{I}{P_y} - \frac{P_x}{P_y}X$$

$$
\text{Slope} = -\frac{P_x}{P_y} \rightarrow \text{opportunity cost of } X \text{ in terms of } Y
$$

$\text{Intercepts:}$
$$X = \frac{I}{P_x}$$
$$Y = \frac{I}{P_y}$$


The budget line shows a trade-off: to buy more of one good, the consumer must give up some of the other.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
income = 100   # changable => 200
Px = 10
Py = 5

# Generate values for X
X = np.linspace(0, income / Px, 100)

# Budget line equation: Y = (I/Py) - (Px/Py)*X
Y = (income / Py) - (Px / Py) * X

# Plot
plt.figure()
plt.plot(X, Y)

# Labels and title
plt.xlabel("Quantity of Good X")
plt.ylabel("Quantity of Good Y")
plt.title("Budget Constraint")

# Axis limits
plt.xlim(0, income / Px)
plt.ylim(0, income / Py)

# Grid
plt.grid()

plt.show()
```

**From the graph:**

If you spend everything on X → you get 10 units

If you spend everything on Y → you get 20 units

**Trade-off:**

- To gain 1 unit of X, you give up 2 units of Y

Because: $$\frac{P_x}{P_y} = \frac{10}{5} = 2$$



2. **Indifference Curves**

An indifference curve shows all combinations of goods that give the consumer the same level of satisfaction (utility).

Key properties:

- Downward sloping
- Higher curves = higher satisfaction
- Convex to origin → diminishing marginal rate of substitution (MRS)
- Curves never cross

They represent the consumer’s preferences.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Budget Constraint ---
income = 100
Px = 10
Py = 5

X = np.linspace(0.1, income / Px, 100)  # avoid division by zero
Y_budget = (income / Py) - (Px / Py) * X

# --- Indifference Curves ---
U_levels = [20, 40, 60]

plt.figure()

# Plot budget line
plt.plot(X, Y_budget, label="Budget Line")

# Plot indifference curves
for U in U_levels:
    Y = U / X
    plt.plot(X, Y, linestyle='--', label=f"U = {U}")

# Labels
plt.xlabel("Quantity of Good X")
plt.ylabel("Quantity of Good Y")
plt.title("Budget Constraint and Indifference Curves")

# Limits
plt.xlim(0, income / Px)
plt.ylim(0, income / Py)

# Legend and grid
plt.legend()
plt.grid()

plt.show()
```

1. Shape
Curves bend inward → diminishing MRS

As you consume more X, you give up less Y

2. Higher Curve = Higher Satisfaction

U = 60 → better than 

U = 40 → better than 

U = 20

3. Affordable vs Unaffordable

- Curves above the budget line → ❌ cannot afford
- Curves below → ✅ affordable but not optimal

4. The Goal

The consumer wants to reach the highest curve possible
BUT still stay on the budget line

Intuition

Budget line = your limit

Indifference curve = your desire

The best choice = highest desire within your limit

3. **The Consumer’s Optimal Choice**

The consumer chooses the combination of goods that:

- Maximizes satisfaction
- While respecting the budget constraint

Condition for optimum:
$$MRS = \frac{P_x}{P_y}$$


The optimal choice occurs where:

The indifference curve is tangent to the budget line

At this point:

- The consumer reaches the highest possible utility
- The entire income is used efficiently

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
income = 100
Px = 10
Py = 5

# X values
X = np.linspace(0.1, income / Px, 100)

# Budget line
Y_budget = (income / Py) - (Px / Py) * X

# Indifference curve at optimal utility
U_opt = 50
Y_opt_curve = U_opt / X

# Optimal point
X_opt = 5
Y_opt = 10

plt.figure()

# Plot budget line
plt.plot(X, Y_budget, label="Budget Line")

# Plot optimal indifference curve
plt.plot(X, Y_opt_curve, linestyle='--', label="Optimal IC (U=50)")

# Plot optimal point
plt.scatter(X_opt, Y_opt)
plt.text(X_opt, Y_opt, "  Optimum (5,10)")

# Labels
plt.xlabel("Quantity of Good X")
plt.ylabel("Quantity of Good Y")
plt.title("Consumer Optimal Choice")

plt.xlim(0, income / Px)
plt.ylim(0, income / Py)

plt.legend()
plt.grid()

plt.show()
```

Key Insight

At the optimal point:

- Budget is fully used ✅
- Utility is maximized ✅
- Slope of Indifference Curve = slope of budget line ✅

That’s why:

Tangency = Optimal Choice

Intuition

Before this point → you can increase satisfaction

After this point → you can’t afford it

So this is the best possible choice

4. **How Changes in Income Affect Choices**

When income changes:

The budget line shifts (parallel shift)

Prices stay constant

- Income ↑ → shift outward
- Income ↓ → shift inward

Result:

The consumer moves to a new optimal point

For most goods (normal goods):

Income ↑ → consumption ↑

For some goods (inferior goods):

Income ↑ → consumption ↓

```python
import numpy as np
import matplotlib.pyplot as plt

# Prices
Px = 10
Py = 5

# Income levels
income1 = 100
income2 = 200

# X values
X = np.linspace(0.1, income2 / Px, 100)

# Budget lines
Y1 = (income1 / Py) - (Px / Py) * X
Y2 = (income2 / Py) - (Px / Py) * X

# Optimal points
X1, Y1_opt = 5, 10
X2, Y2_opt = 10, 20

plt.figure()

# Plot budget lines
plt.plot(X, Y1, label="Income = 100")
plt.plot(X, Y2, label="Income = 200")

# Plot optimal points
plt.scatter(X1, Y1_opt)
plt.text(X1, Y1_opt, "  A (5,10)")

plt.scatter(X2, Y2_opt)
plt.text(X2, Y2_opt, "  B (10,20)")

# Labels
plt.xlabel("Quantity of Good X")
plt.ylabel("Quantity of Good Y")
plt.title("Effect of Income Increase")

plt.xlim(0, income2 / Px)
plt.ylim(0, income2 / Py)

plt.legend()
plt.grid()

plt.show()
```

Parallel Shift

- Slope unchanged → prices unchanged
- Only purchasing power increases

Higher Income → Higher Utility

Old utility:


U = 5⋅10=50

New utility:

U = 10⋅20=200

Consumer is better off ✅

Type of Goods

In this example:

- X ↑ when income ↑
- Y ↑ when income ↑

👉 Both are normal goods

🔹 7. Intuition (Simple)

Income increase = “You can afford more of everything”

So:

Budget line moves outward

You reach a higher indifference curve

You consume more

5. **How Changes in Prices Affect Choices**

When the price of a good changes:

The budget line rotates
- Price ↓ → rotates outward
- Price ↑ → rotates inward

Result:

A new optimal choice is made

This change can be divided into:

Substitution Effect

The consumer replaces the relatively expensive good with the cheaper one.

Income Effect

Depends on type of good:
- Normal → reinforces substitution
- Inferior → works opposite


```python
import numpy as np
import matplotlib.pyplot as plt

# Income
income = 100

# Prices
Px_old = 10
Px_new = 5
Py = 5

# X values
X = np.linspace(0.1, 20, 100)

# Budget lines
Y_old = (income / Py) - (Px_old / Py) * X
Y_new = (income / Py) - (Px_new / Py) * X

# Optimal points
X_old, Y_old_opt = 5, 10
X_new, Y_new_opt = 10, 10

plt.figure()

# Plot budget lines
plt.plot(X, Y_old, label="Old Price (Px=10)")
plt.plot(X, Y_new, label="New Price (Px=5)")

# Plot optimal points
plt.scatter(X_old, Y_old_opt)
plt.text(X_old, Y_old_opt, "  A (5,10)")

plt.scatter(X_new, Y_new_opt)
plt.text(X_new, Y_new_opt, "  B (10,10)")

# Labels
plt.xlabel("Quantity of Good X")
plt.ylabel("Quantity of Good Y")
plt.title("Effect of Price Change")

plt.xlim(0, 20)
plt.ylim(0, 20)

plt.legend()
plt.grid()

plt.show()
```

Key Insights

Rotation (Not Shift)

- Only one price changed
- Slope changes

Law of Demand

Price ↓ → Quantity demanded ↑

You moved from:

X = 5 → X = 10

Two Effects Together

Total change = Substitution Effect + Income Effect

Intuition

X becomes cheaper → you switch toward it

You feel richer → you buy more

Both push you toward more X

6. **Deriving the Demand Curve**

The demand curve is derived from the consumer’s optimal choices as price changes.

Process:

(a) Change the price of a good

(b) Observe the new optimal quantity chosen

(c) Record price–quantity combinations

These combinations form the demand curve:

- Price ↓ → quantity demanded ↑
- Price ↑ → quantity demanded ↓

Thus, the demand curve reflects how rational consumer choices respond to price changes.

**The Idea**

We repeatedly:

Change the price of X
Find the optimal quantity of X
Record the result

Then we plot:

Price vs Quantity → Demand Curve

**Use Our Model (Very Important)**

We keep:
$$
\text{Utility: } U = X \cdot Y \\
\text{Income: } I = 100 \\
P_y = 5 \\
\text{We vary only } P_x
$$
	

**Solve Generally (Powerful Insight)**

From optimal condition:

$$
\frac{Y}{X} = \frac{P_x}{P_y} \Rightarrow Y = \frac{P_x}{P_y} X
$$

Substitute into budget:

$$
\begin{aligned}
P_x X + P_y \left( \frac{P_x}{P_y} X \right) &= I \\
P_x X + P_x X &= I \\
2P_x X &= I \\
X &= \frac{I}{2P_x}
\end{aligned}
$$


Demand Function

$$
X^* = \frac{100}{2P_x} = \frac{50}{P_x}
$$

	​
This is your demand curve equation

**Create Dataset**

Let’s vary price:

$$
\begin{array}{|c|c|}
\hline
\textbf{P}_x & \textbf{X} = 50 / P_x \\ \hline
2 & 25 \\ \hline
5 & 10 \\ \hline
10 & 5 \\ \hline
20 & 2.5 \\ \hline
\end{array}
$$

You can already see:

Price ↓ → Quantity ↑ ✅

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
income = 100

# Prices of X
Px_values = np.array([2, 5, 10, 20])

# Demand function: X = 50 / Px
X_values = 50 / Px_values

plt.figure()

# Plot demand curve
plt.plot(X_values, Px_values, marker='o')

# Labels
plt.xlabel("Quantity of Good X")
plt.ylabel("Price of Good X")
plt.title("Demand Curve")

plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
income1 = 100
income2 = 200
Px1 = 10
Px2 = 5
Py = 5

# X range
X = np.linspace(0.1, 20, 200)

# Budget lines
Y_budget1 = (income1 / Py) - (Px1 / Py) * X
Y_budget2 = (income2 / Py) - (Px1 / Py) * X   # income change
Y_budget3 = (income1 / Py) - (Px2 / Py) * X   # price change

# Indifference curves
U_levels = [30, 50, 80]
IC_curves = [U / X for U in U_levels]

# Optimal points
A = (5, 10)    # original
B = (10, 20)   # income increase
C = (10, 10)   # price decrease

plt.figure()

# Plot budget lines
plt.plot(X, Y_budget1, label="Initial Budget")
plt.plot(X, Y_budget2, linestyle='--', label="Income Increase")
plt.plot(X, Y_budget3, linestyle=':', label="Price Decrease")

# Plot indifference curves
for i, U in enumerate(U_levels):
    plt.plot(X, IC_curves[i], linestyle='-.', label=f"U={U}")

# Plot optimal points
plt.scatter(*A)
plt.text(*A, "  A")

plt.scatter(*B)
plt.text(*B, "  B")

plt.scatter(*C)
plt.text(*C, "  C")

# Labels
plt.xlabel("Good X")
plt.ylabel("Good Y")
plt.title("Consumer Choice: Budget & Indifference Curves")

plt.xlim(0, 20)
plt.ylim(0, 40)

plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------------------------
# 1. Crossing Indifference Curve (Cobb-Douglas U = X*Y)
# ----------------------------------------------------------------------
X = np.linspace(0.1, 12, 200)

# Indifference curve: U = 30
U_cross = 30
Y_cross = U_cross / X

# Budget line: 4X + 4Y = 40
income = 40
Px = 4
Py = 4
Y_budget = (income / Py) - (Px / Py) * X

# Find intersection points (solving X*Y = 30 and 4X+4Y=40)
# From budget: X + Y = 10 => Y = 10 - X
# Substitute: X*(10 - X) = 30 => -X^2 + 10X - 30 = 0 => X^2 - 10X + 30 = 0
# Roots: X = 5 ± sqrt(25 - 30) = 5 ± sqrt(-5) → no real intersection? Wait, that's wrong.
# Let's solve correctly: 4X + 4Y = 40 → X + Y = 10 → Y = 10 - X.
# Then U = X*(10 - X) = 10X - X^2 = 30 → X^2 - 10X + 30 = 0.
# Discriminant = 100 - 120 = -20 → no real intersection! So the indifference curve U=30 does not touch the budget line.
# To have crossing, choose a different U level. Let's pick U = 25.
U_cross = 25
Y_cross = U_cross / X
# Now solve X*(10 - X) = 25 → X^2 - 10X + 25 = 0 → (X-5)^2=0 → X=5, Y=5.
intersect_X = 5
intersect_Y = 5

plt.figure(figsize=(10, 6))
plt.plot(X, Y_cross, '--', label=f'Indifference Curve (U={U_cross})')
plt.plot(X, Y_budget, '-', label='Budget Line (4X+4Y=40)')
plt.scatter(intersect_X, intersect_Y, color='red', label='Tangency / Crossing Point')

# ----------------------------------------------------------------------
# 2. Corner Solution (Perfect Substitutes U = X + Y)
# ----------------------------------------------------------------------
# Utility: U = X + Y. With Px = 1, Py = 10, income = 40.
# The consumer will spend all income on the cheaper good (X) -> corner at (40,0).
income_corner = 40
Px_corner = 1
Py_corner = 10
X_corner = np.linspace(0, 40, 200)
Y_budget_corner = (income_corner / Py_corner) - (Px_corner / Py_corner) * X_corner
plt.plot(X_corner, Y_budget_corner, '-', color='green', label='Budget Line (Corner: 1X+10Y=40)')
plt.scatter(income_corner / Px_corner, 0, color='orange', label='Optimal Corner (X only)')

# Labels and formatting
plt.xlabel('Good X')
plt.ylabel('Good Y')
plt.title('Crossing Indifference Curve & Corner Solution')
plt.legend()
plt.grid()
plt.xlim(0, 40)
plt.ylim(0, 12)
plt.show()
```
