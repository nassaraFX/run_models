### 1. **Understanding Demand**

#### (a) **The Meaning of Demand**

In economics, the term demand has a very specific meaning. It is far more than just a desire or a wish for a product. A person might want a luxury yacht, but that want is not considered demand in the economic sense.

For economists, demand refers to the quantity of a good or service that consumers are both willing and able to purchase at various prices during a given period of time.

This definition rests on **two crucial pillars**:

- ***Willingness to Buy*** The consumer must desire the product.

- ***Ability to Pay*** The consumer must possess the necessary purchasing power (income) to acquire it.

Only when both conditions are met can we speak of effective demand.

#### (b) **The Law of Demand**

As you correctly identified, the relationship between price and quantity demanded is predictable. This relationship is formalized in economics as the Law of Demand. This law states that, all other factors being equal (**ceteris paribus**), as the price of a good or service increases, the quantity demanded decreases, and conversely, as the price decreases, the quantity demanded increases.

The relationship is, therefore, inverse or negative. Price (P) and Quantity Demanded (Qd) move in opposite directions.

**Why does this law hold true?**

**Two primary reasons explain this consistent behavior:**

***The Substitution Effect*** When the price of a good rises, it becomes more expensive relative to its substitutes. Consumers will naturally switch to alternative products that can satisfy the same need but are now cheaper. For example, if the price of coffee increases significantly, consumers may start buying more tea.

***The Income Effect*** When the price of a good rises, a consumer's purchasing power is effectively reduced. Their income can now buy fewer goods and services than before. This makes them feel poorer, leading them to reduce their consumption of this now more expensive good, along with possibly other non-essential items.

#### (c) **The Demand Schedule**

A demand schedule is a table that shows the relationship between the price of a good and the quantity demanded. It is a numerical representation of the law of demand.

Let's say a ***juice seller***

Demand Schedule for Bottles of Juice

| Price (P) per Bottle | Quantity Demanded (Qd) per Week |
| :--- | :--- |
| $2 | 100 |
| $4 | 80 |
| $6 | 60 |
| $8 | 40 |
| $10 | 20 |

**Observation can be done, if the dataset is not large** So this table clearly illustrates the law of demand. At the lowest price of $2, consumers are willing and able to buy 100 bottles. At the highest price of $10, the quantity demanded falls to just 20 bottles. As price rises, quantity demanded consistently falls.

#### (d) **The Demand Curve**

The information in a demand schedule can be transferred to a graph to create a demand curve. The demand curve is a visual representation of the law of demand.

- The vertical ***axis (y-axis)*** measures the Price (P) of the good.

- The horizontal ***axis (x-axis)*** measures the Quantity Demanded (Qd) .

When we plot the points from the demand schedule above and connect them, we get a curve that slopes downward from left to right. This negative slope is the graphical signature of the law of demand.

"A graph would be inserted here showing a downward-sloping line labeled 'D', with Price on the Y-axis and Quantity Demanded on the X-axis, and the points from the schedule plotted."

#### (e) **Distinction Between Demand and Quantity Demanded**

One of the most important concepts to master is the difference between a change in "demand" and a change in "quantity demanded." They are not the same thing.

 1. **Change in Quantity Demanded:**

***Definition:*** A movement along a fixed demand curve.

***Cause:*** It is caused exclusively by a change in the price of the good itself.

***Example:*** In our schedule, a price change from ***$4*** to ***$6*** causes the quantity demanded to move from ***80*** to ***60***. We are moving from one point to another on the same curve.

 2. **Change in Demand:**

***Definition:*** A shift of the entire demand curve to the left (***decrease***) or to the right (***increase***).

***Cause:*** It is caused by a change in any factor other than the good's own price. These are the "***all other factors***" we held equal (*ceteris paribus*) for the law of demand.

 3. **Examples of factors that shift demand:**

***Changes in Consumer Income:*** If consumer incomes ***rise***, the demand for most goods (*called normal goods*) will ***increase***, shifting the curve to ***the right***. At every price, more is demanded.

***Changes in Tastes and Preferences:*** A successful advertising campaign or a ***positive*** health report can ***increase demand for a product***, shifting the curve ***rightward***. A ***negative*** report would shift it ***leftward***.

 3. **Changes in the Price of Related Goods:**

*Substitutes (e.g., tea and coffee):* If the price of tea ***rises***, the demand for coffee (*its substitute*) will ***increase***, shifting the coffee demand curve to ***the right***.

*Complements (e.g., printers and ink cartridges):* If the price of printers ***falls***, more people buy printers, which ***increases the demand for ink cartridges*** (t*heir complement*), shifting the ink demand curve to ***the right***.

*Changes in Expectations:* If consumers expect the price of a good to ***rise*** in the future (e.g., a sale ending), they may buy more of it now, ***increasing*** current demand (curve shifts *right*).

*Changes in the Number of Buyers:* An **increase in population** in the market area will ***increase the demand for most goods***, shifting the curve to the ***right***.

***In summary:*** A price change causes a movement along the curve. A change in anything else that influences buying plans causes the entire curve to shift.

```python
# Demand data
prices = [2, 4, 6, 8, 10]
quantity_demanded = [100, 80, 60, 40, 20]

demand_df = pd.DataFrame({
    'Price ($)': prices,
    'Quantity Demanded': quantity_demanded
})

print("Demand Schedule for Juice Bottles")
print(demand_df.to_string(index=False))
print("\n" + "=" * 32 + "\n")

# Increased demand schedule (shift to the right)
shifted_quantity = [q + 20 for q in quantity_demanded]

# Create figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# LEFT: Movement along demand curve
ax1.plot(
    quantity_demanded, prices,
    marker='o',
    linestyle='-',
    linewidth=2.5,
    markersize=8,
    label='Demand Curve (D)',
)

# Fill area lightly for attractiveness
ax1.fill_between(quantity_demanded, prices, alpha=0.1)

# Points for movement
point1 = (80, 4)
point2 = (60, 6)

ax1.scatter(*point1, s=80, zorder=3)
ax1.scatter(*point2, s=80, zorder=3)

ax1.annotate(
    '',
    xy=point2,
    xytext=point1,
    arrowprops=dict(arrowstyle='->', lw=2)
)

ax1.text(
    68, 5.2,
    'Price rises\nQuantity demanded falls',
    fontsize=10,
    ha='center',
    bbox=dict(boxstyle='round,pad=0.4', alpha=0.2)
)

ax1.set_xlabel('Quantity Demanded (bottles per week)', fontsize=11)
ax1.set_ylabel('Price ($ per bottle)', fontsize=11)
ax1.set_title('Movement Along the Demand Curve', fontsize=13, fontweight='bold')
ax1.legend()
ax1.set_xlim(0, 120)
ax1.set_ylim(0, 12)


# RIGHT: Shift in demand
ax2.plot(
    quantity_demanded, prices,
    marker='o',
    linestyle='-',
    linewidth=2.5,
    markersize=8,
    label='Original Demand (D0)'
)

ax2.plot(
    shifted_quantity, prices,
    marker='s',
    linestyle='--',
    linewidth=2.5,
    markersize=8,
    label='Increased Demand (D1)'
)

# Arrow showing rightward shift
ax2.annotate(
    '',
    xy=(shifted_quantity[2], prices[2]),
    xytext=(quantity_demanded[2], prices[2]),
    arrowprops=dict(arrowstyle='->', lw=2)
)

ax2.text(
    67, 6.5,
    'Demand shifts right\nbecause of a non-price factor',
    fontsize=10,
    ha='left',
    bbox=dict(boxstyle='round,pad=0.4', alpha=0.2)
)

ax2.set_xlabel('Quantity Demanded (bottles per week)', fontsize=11)
ax2.set_ylabel('Price ($ per bottle)', fontsize=11)
ax2.set_title('Increase in Demand (Shift Right)', fontsize=13, fontweight='bold')
ax2.legend()
ax2.set_xlim(0, 140)
ax2.set_ylim(0, 12)

# Main title
fig.suptitle('Movement vs. Shift', fontsize=16, fontweight='bold')

# Cleaner layout
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

**The right graph shows a parallel shift by adding 20 to every quantity. This is correct for a simple classroom example, although in reality demand does not always shift by exactly the same amount at every price**

### 2. **Demand Function**

While graphs provide an excellent visual representation of demand, economists often need a more precise tool to analyze and predict consumer behavior. This tool is the demand equation (also called the demand function). It expresses the relationship between quantity demanded and price in a mathematical formula.

For simplicity, we often assume that the relationship between price and quantity demanded is a straight line. This is called a linear demand function. Its general form is

$Q = a + bP$


- $Q$ represents the quantity demanded.

- $P$ represents the price of the good.

- $a$ is the intercept term.

- $b$ is the slope coefficient.

A More Complete Demand Function
To capture the effect of non‑price factors directly, economists often write a multivariable demand function. Instead of $Q = a + bP$, we might have:

$Q = a + bP + cI + dP_s + eP_c$

$I$ = consumer income

$P_s$ = price of substitutes

$P_c$ = price of complements

$c, d, e$ = coefficients that measure how each factor affects quantity demanded.


Because the Law of Demand states that price and quantity demanded are inversely related, the coefficient $b$ will always be negative ($b < 0$).

#### (a) **The Intercept ($a$)**

The intercept, $a$, is the value of $Q$ when the price $P$ is equal to zero.

$Q = a + b(0) = a$

In our ongoing juice example, let's consider the demand equation:

$Q = 120 − 10P$

Here, the intercept $a$ is 120.

***Interpretation:*** If the juice were given away for free ($P=0$), the quantity demanded would be 120 bottles per week.

It is crucial to understand that the intercept is often a theoretical concept. In the real world, a price of zero is rare, and the demand at that price might be different due to factors like rationing or storage costs. However, mathematically, the intercept serves as the anchor point for the demand line on the graph—it is the point where the line would hit the quantity axis (the x-axis) if we extended it far enough.

#### (b) **The Slope ($b$)**

The slope, $b$, is the most informative part of the equation. It measures the responsiveness of quantity demanded to a change in price. More formally, it tells us how many units $Q$ changes when $P$ changes by one unit.

Using our equation $Q = 120 - 10P$, the slope $b$ is -10.

***Interpretation:*** For every one-unit increase in price, the quantity demanded decreases by 10 units.

Let's test this with our original demand schedule:

At a price of $4$: $Q = 120 - (10 \times 4) = 120 - 40 = 80$ (Matches the schedule)

If price increases by one unit to $5$: $Q = 120 - (10 \times 5) = 120 - 50 = 70$

As predicted by the slope, the quantity fell from 80 to 70—a decrease of exactly 10 units.

The slope is the "engine" of the demand function, quantifying the sensitivity of consumers to price changes. A larger negative slope (e.g., -20) would mean demand is more sensitive, dropping by 20 units for every $1 price increase. A smaller negative slope (e.g., -2) would mean demand is less sensitive.

#### (c) **Deriving the Demand Equation from the Schedule**

We can derive the linear demand equation directly from the demand schedule. Let's use two points from our original table: ($Q=100$, $P=2$) and ($Q=80$, $P=4$).

Calculate the slope ($b$). The slope is the "rise" over the "run," but here the "rise" is the change in $Q$ and the "run" is the change in $P$.

$b = \frac{\Delta Q}{\Delta P} = \frac{80 - 100}{4 - 2} = \frac{-20}{2} = -10$

Find the intercept ($a$). Now that we know $b = -10$, we can plug one of the points (e.g., $Q=100$, $P=2$) into the general equation $Q = a + bP$ to solve for $a$.

\begin{aligned}
100 &= a + (-10)(2) \\
100 &= a - 20 \\
a &= 100 + 20 \\
a &= 120
\end{aligned}


Write the equation. Substituting $a = 120$ and $b = -10$ into the general form gives us:
$Q = 120 − 10P$

This is the precise mathematical representation of the demand curve we plotted earlier. Every point on that line satisfies this equation.

#### (d) **The Inverse Demand Equation**

Sometimes, it is useful to express price as a function of quantity. This is called the inverse demand function. We can derive it by simply rearranging our original equation to solve for $P$.

Starting with $Q = 120 - 10P$:

Add $10P$ to both sides: $10P + Q = 120$

Subtract $Q$ from both sides: $10P = 120 - Q$

Divide both sides by 10: 

$P = 12 − 0.1Q$

In this form, the equation tells us the maximum price consumers are willing to pay for a given quantity. For example, to sell 60 bottles ($Q=60$), the highest price we can charge is $P = 12 - (0.1 \times 60) = 12 - 6 = 6$, which matches our original schedule.

The demand equation is a powerful tool. It allows us to calculate exact quantities at any given price, measure consumer sensitivity through the slope, and forms the foundation for more advanced economic analysis, such as calculating consumer surplus and market equilibrium.



### 1. **Simple Linear Model**

Understanding Its Core Properties

**What Does "Simple Linear Model" Really Mean?**

When economists refer to a simple linear model, they are making a specific mathematical assumption about the relationship between two variables. The general form is:

$Y = a + bX$

In the context of demand, we write:

$Q = a + bP$


The term ***"linear"*** indicates that when we plot this relationship on a graph, it forms a straight line. However, the practical meaning extends beyond the visual representation.

The Defining Characteristic: Constant Additive Change

The essential property of a linear model is that every one-unit increase in the independent variable ($P$) changes the dependent variable ($Q$) by exactly the same amount. This constant amount is the slope coefficient $b$.

Consider our demand equation from the previous section:

\begin{aligned}
b &= \frac{\Delta Q}{\Delta P} = -10 \\
100 &= a + (-10)(2) \\
a &= 120 \\
Q &= 120 - 10P
\end{aligned}


Let us examine what happens as price increases by 1 cedi at a time

| Price Change (GH₵) | Calculation ($Q = 120 - 10P$) | Change in Quantity | Total Revenue ($P \times Q$) |
| :--- | :--- | :--- | :--- |
| From 2 to 3 | $Q_2 = 120 - 10(2) = 100$ <br> $Q_3 = 120 - 10(3) = 90$ | Decrease of 10 units | $TR_2 = 200$ <br> $TR_3 = 270$ |
| From 3 to 4 | $Q_3 = 90$ <br> $Q_4 = 120 - 10(4) = 80$ | Decrease of 10 units | $TR_3 = 270$ <br> $TR_4 = 320$ |
| From 4 to 5 | $Q_4 = 80$ <br> $Q_5 = 120 - 10(5) = 70$ | Decrease of 10 units | $TR_4 = 320$ <br> $TR_5 = 350$ |
| From 7 to 8 | $Q_7 = 120 - 10(7) = 50$ <br> $Q_8 = 120 - 10(8) = 40$ | Decrease of 10 units | $TR_7 = 350$ <br> $TR_8 = 320$ |
| From 9 to 10 | $Q_9 = 120 - 10(9) = 30$ <br> $Q_{10} = 120 - 10(10) = 20$ | Decrease of 10 units | $TR_9 = 270$ <br> $TR_{10} = 200$ |


Observe the pattern: whether price is low (moving from 2 to 3) or high (moving from 7 to 8), the change in quantity is invariably 10 units. This phenomenon is known as an additive change or a constant marginal effect.

Mathematically, we express this property as:

$$\frac{\Delta Q}{\Delta P} = b \text{ (constant)}$$

The slope $b$ is constant at every point along the line. This constancy is both the strength and the limitation of the linear model.

#### **Practical Prediction with the Simple Linear Model**

Because the linear model exhibits a consistent, predictable relationship, it becomes an invaluable tool for forecasting and decision-making. Once we have estimated the parameters $a$ (intercept) and $b$ (slope), we can predict the quantity demanded at any given price with simple arithmetic.

Predicting Demand at Various Prices

Using our demand function $Q = 120 - 10P$

* **If the firm sets price at 3 cedis:**

  $Q = 120 - 10(3) = 120 - 30 = 90 \text{ units}$

* **If the firm sets price at 5 cedis:**

  $Q = 120 - 10(5) = 120 - 50 = 70 \text{ units}$

* **If the firm sets price at 7 cedis:**

  $Q = 120 - 10(7) = 120 - 70 = 50 \text{ units}$

* **If the firm sets price at 9 cedis:**

  $Q = 120 - 10(9) = 120 - 90 = 30 \text{ units}$


Managerial Applications

The linear demand model enables managers to answer strategic questions with precision:

**What is the sales impact of a price change?**

If price increases by 2 cedis, quantity demanded falls by 20 units (since $\Delta Q = b \times \Delta P = -10 \times 2 = -20$).

**What price should be charged to achieve a target sales volume?**

To sell 60 units, solve $60 = 120 - 10P$:

$
\begin{aligned}
10P &= 120 - 60 \\
10P &= 60 \\
P &= 6 \text{ cedis}
\end{aligned}
$


**How much must price change to achieve a desired increase in sales?**

To increase sales by 20 units, we need $\Delta Q = 20$. Since $\Delta Q = b \times \Delta P$, we have:

$
\begin{aligned}
20 &= -10 \times \Delta P \\
\Delta P &= \frac{20}{-10} \\
\Delta P &= -2 \text{ cedis}
\end{aligned}
$

The firm must lower price by 2 cedis.

The linear model thus transforms the abstract concept of demand into a concrete, calculable tool for business analysis.

**Why the Simple Linear Model Is Sometimes Not Enough?**

Despite its usefulness and simplicity, the linear model rests on an assumption that may not hold in many real-world markets. The limitation stems from the nature of consumer decision-making.

The Problem with Additive Changes

Recall that in a linear model, a 1-cedi price increase always reduces quantity by the same absolute amount (10 units in our example). However, consider what this means in percentage terms:


| Starting Price | 1-Cedi Increase | Percentage Price Increase | Absolute Q Drop | Percentage Q Drop |
| :--- | :--- | :--- | :--- | :--- |
| 2 cedis | to 3 cedis | 50% increase | 10 units | 10% drop (from 100 to 90) |
| 8 cedis | to 9 cedis | 12.5% increase | 10 units | 25% drop (from 40 to 30) |

Notice the inconsistency: the same absolute price increase represents vastly different percentage increases depending on the starting price. At low prices (2 cedis), a 1-cedi increase is a substantial 50% price hike. At high prices (8 cedis), the same 1-cedi increase is merely a 12.5% rise.

Yet the linear model assumes consumers react identically in both situations—reducing purchases by exactly 10 units regardless of whether the price rose by 50% or 12.5%. Is this psychologically plausible?

### 2. **Log-Log Model**

(a) **Introducing the Log-Log Model**

We have seen that the linear model assumes consumers respond to absolute changes in price. However, as discussed in Step 9, real-world consumers often think in terms of percentages. To capture this behavior, economists use the log-log model (also called the constant elasticity model).

The general form of a log-log model is:

$$
\ln Y = a + b \ln X
$$


For demand, we write:

$$
\ln Q = a + b \ln P
$$


$\ln Q$ represents the natural logarithm of quantity demanded

$\ln P$ represents the natural logarithm of price

$a$ is the intercept term (in log form)

$b$ is the slope coefficient, which has a special interpretation

***Important Note:*** The notation $\ln$ refers to the natural logarithm (logarithm to the base $e$, where $e \approx 2.71828$). In economics, when we say "log" without specifying the base, we almost always mean the natural logarithm.

(b) **The Meaning of Slope in the Log-Log Model**

The interpretation of the slope coefficient $b$ in a log-log model is fundamentally different from that in a linear model. This difference is crucial for understanding and applying the model correctly.

In the linear model:

Slope means: a 1-unit increase in $P$ changes $Q$ by $b$ units

In the log-log model:

Slope means: a 1% increase in $P$ changes $Q$ by $b%$

This is because of a mathematical property of logarithms: when variables are expressed in natural logs, the coefficient measures the elasticity—the percentage change in the dependent variable resulting from a 1% change in the independent variable.

Illustrative Example

Consider the following log-log demand function:

$$
\ln Q = 5 - 0.8 \ln P
$$

Here, the slope coefficient is:

$$
b = -0.8
$$

***Interpretation:*** If price increases by 1%, quantity demanded decreases by 0.8%.

This coefficient $b$ is given a special name: the price elasticity of demand. It is one of the most important concepts in economics.

Formal Definition of Price Elasticity

The price elasticity of demand ($E_p$) is defined as:

$$
E_p = \frac{\text{Percentage change in quantity demanded}}{\text{Percentage change in price}} = \frac{\%\Delta Q}{\%\Delta P}
$$

In a log-log model, this elasticity is constant at all points along the demand curve—hence the name "constant elasticity model."

(c) **Why This Is Different from the Linear Model**
To appreciate the distinction, let us place the two models side by side.

Linear Model:

$$
Q = 120 - 10P
$$

***Interpretation:*** Price increases by 1 unit → Quantity decreases by 10 units

This is a fixed unit change (additive effect)

The effect is measured in absolute terms

Log-Log Model:

$$
\ln Q = 5 - 0.8 \ln P
$$

***Interpretation:*** Price increases by 1% → Quantity decreases by 0.8%

This is a fixed percentage change (multiplicative effect)

The effect is measured in relative (percentage) terms

The fundamental difference can be summarized as:

| Model | Type of Change | What Is Constant |
| :--- | :--- | :--- |
| **Linear** | Additive (unit change) | The absolute change in $Q$ per unit change in $P$ |
| **Log-Log** | Multiplicative (percentage change) | The percentage change in $Q$ per percentage change in $P$ |

(e) **The Multiplicative Nature of the Log-Log Model**

Because the log-log model operates in percentages, it is often described as multiplicative rather than additive. To understand this, let us rewrite the log-log model in its exponential form.

***Starting from:***

$$
\ln Q = a + b \ln P
$$


***We can exponentiate both sides:***

$$
\begin{aligned}
e^{\ln Q} &= e^{a + b \ln P} \\
Q &= e^a \times e^{b \ln P} \\
Q &= e^a \times (e^{\ln P})^b \\
Q &= A \times P^b
\end{aligned}
$$

Where $A = e^{a}$ is a constant.

***The Multiplicative Form:***

$$
Q = A P^b
$$

***Notice the structure:***

***In the linear model:*** $Q = a + bP$ (terms are added)

***In the log-log model:*** $Q = A \times P^b$ (terms are multiplied)

This multiplicative structure means that a change in price has a proportional effect on quantity. If price doubles ($P$ is multiplied by 2), quantity is multiplied by $2^b$. Since $b$ is negative (law of demand), quantity decreases by a proportional amount.

***Example:*** If $b = -0.8$ and price doubles ($P$ multiplied by 2), then:

$$
\begin{aligned}
Q_{new} &= A \times (2P)^b \\
Q_{new} &= A \times 2^b \times P^b \\
Q_{new} &= 2^b \times Q_{original}
\end{aligned}
$$

**Using your specific elasticity ($b = -0.8$):**

$$
2^{-0.8} \approx 0.574
$$

So quantity falls to about 57.4% of its original value—a decrease of 42.6%.


#### **Summary**

| Feature | Linear Model | Log-Log Model |
| :--- | :--- | :--- |
| **Equation form** | $Q = a + bP$ | $\ln Q = a + b \ln P$ or $Q = AP^b$ |
| **Nature of relationship** | Additive | Multiplicative |
| **Interpretation of slope** | $\Delta Q$ for a 1-unit change in $P$ | $\% \Delta Q$ for a 1% change in $P$ (elasticity) |
| **Effect of price change** | Same absolute change at all prices | Same percentage change at all prices |
| **Graph shape (P-Q axes)** | Straight line | Curve convex to origin |
| **Graph shape (ln P-ln Q)** | Curve | Straight line |
| **Elasticity** | Varies along the curve | Constant along the curve |
| **When to use** | Small price changes, narrow ranges | Large changes, wide ranges, constant elasticity |

#### 3. **Curved Shape of the Log-Log Model**

A question often arises when students first encounter the log-log model: Why do economists say that the log-log model represents a curve when plotted on a standard graph?

The answer lies in the fundamental difference between additive and multiplicative relationships.

**Linear Model:** A Straight Line

Recall that a linear demand equation, such as $Q = 120 - 10P$, produces a straight line when plotted on a graph with $P$ on the vertical axis and $Q$ on the horizontal axis. This is because the relationship is additive: each one-unit increase in $P$ reduces $Q$ by exactly the same amount (10 units) regardless of where we are on the line. The slope is constant, and the graph is a straight line.

**Log-Log Model:** A Curve

A log-log model, such as $Q = AP^b$, produces a curve when plotted on the same standard axes. Why? Because the effect of price changes is based on percentages, not fixed units.

Consider what happens as price increases:

***At low prices:*** A given absolute price increase represents a large percentage change. Since quantity responds by a fixed percentage, the absolute drop in quantity will be large.

***At high prices:*** The same absolute price increase now represents a small percentage change. Consequently, the absolute drop in quantity will be smaller.

This pattern—large absolute drops at low prices, smaller absolute drops at high prices—creates a curve that is convex to the origin (bending toward the axes).

Visualizing the Curvature

Imagine a typical log-log demand curve:

Starting from a very low price, an increase of 1 cedi might cause quantity to fall by 20 units

At a medium price, the same 1-cedi increase might cause quantity to fall by only 10 units

At a high price, a 1-cedi increase might cause quantity to fall by just 5 units

The slope is not constant; it becomes steeper (in absolute terms) at low prices and flatter at high prices. This changing slope is what gives the curve its characteristic curved shape.

**Practical Market Example:** Cassava Demand

Let us ground this discussion in a concrete example from an actual market. Consider the demand for cassava, a staple food in many countries. Suppose we observe the following data:

Cassava Demand Data

| Price (P) in cedis | Quantity Demanded (Q) |
| :--- | :--- |
| 1 | 100 |
| 2 | 70 |
| 4 | 50 |
| 8 | 35 |

What do we observe?

Look carefully at what happens each time price doubles:


| Price Change | Price Doubles From → To | Quantity Change |
| :--- | :--- | :--- |
| **First doubling** | 1 → 2 | $100 \rightarrow 70$ (drop of 30 units) |
| **Second doubling** | 2 → 4 | $70 \rightarrow 50$ (drop of 20 units) |
| **Third doubling** | 4 → 8 | $50 \rightarrow 35$ (drop of 15 units) |

Key Observation: The quantity is not falling by the same number of units each time. The absolute drops (30, 20, 15) are getting smaller as price increases.

This pattern is typical of many real-world markets. At very low prices, consumers are highly responsive to price changes. At higher prices, only the most dedicated consumers remain, and their response to further price increases is more muted in absolute terms.

***Why a Linear Model Would Fail?***

If we tried to fit a straight line through these points, we would find that the line misses most of them. A linear model assumes constant unit drops, but our data shows decreasing unit drops. The linear model would systematically overpredict quantity at some prices and underpredict at others.

***Why a Log-Log Model Might Fit Better?***

The pattern we observe—consistent proportional responses—suggests that a log-log model could represent this demand well. Let us check if the percentage changes are roughly constant:

| Price Doubling | Percentage Price Increase | Quantity Change | Approximate % Quantity Drop |
| :--- | :--- | :--- | :--- |
| **1 → 2** | 100% | $100 \rightarrow 70$ | 30% |
| **2 → 4** | 100% | $70 \rightarrow 50$ | 28.6% |
| **4 → 8** | 100% | $50 \rightarrow 35$ | 30% |

The percentage drops are approximately constant (around 29-30%), even though the absolute drops vary considerably. This is exactly the kind of behavior that the log-log model captures.

***How to Move from Normal Numbers to Logs?***

Now that we suspect a log-log model might be appropriate for our cassava data, how do we actually estimate such a model? The process involves transforming the data by taking natural logarithms.

**Step 1:** Start with the original data

| Price (P) | Quantity (Q) |
| :--- | :--- |
| 1 | 100 |
| 2 | 70 |
| 4 | 50 |
| 8 | 35 |

**Step 2:** Take the natural logarithm of each value

Using a calculator or statistical software, we compute $\ln P$ and $\ln Q$:

| Price (P) | Quantity (Q) | $\ln(P)$ | $\ln(Q)$ |
| :--- | :--- | :--- | :--- |
| 1 | 100 | 0.000 | 4.605 |
| 2 | 70 | 0.693 | 4.249 |
| 4 | 50 | 1.386 | 3.912 |
| 8 | 35 | 2.079 | 3.555 |

**Note:** The natural logarithm of 1 is 0. The natural logarithm of $e$ (approximately 2.718) is 1. These values are calculated using the fact that if $y = e^x$, then $\ln y = x$.

**Step 3:** Fit a linear relationship on the transformed data

Instead of fitting the original linear model:
$$Q = a + bP$$

We now fit a ***linear model on the logs***:
$$\ln Q = a + b \ln P$$

This is the essence of the log process: we take logs of both variables, then run a linear regression on the transformed data. The relationship between $\ln Q$ and $\ln P$ is now **linear**, even though the relationship between the original $Q$ and $P$ is **curved**.

**Step 4:** Interpret the slope as elasticity

When we estimate this model, the coefficient **$b$** is the **price elasticity of demand**. For our cassava data, we can calculate the slope between the first and last points:

$$
\begin{aligned}
b &= \frac{\ln Q_4 - \ln Q_1}{\ln P_4 - \ln P_1} \\
b &= \frac{3.555 - 4.605}{2.079 - 0.000} \\
b &= \frac{-1.05}{2.079} \approx -0.505
\end{aligned}
$$

**Interpretation:** This suggests that, on average, a **1% increase in price** leads to about a **0.505% decrease** in quantity demanded.

### **Estimating Demand**

(a) **Introduction to Regression**

From Theory to Data

In the previous part, we developed theoretical models of demand:

**The linear model:** $Q = a + bP$

**The log-log model:** $\ln Q = a + b \ln P$ (or $Q = AP^b$)

These models describe how quantity demanded should behave according to economic theory. However, in the real world, we do not know the true values of $a$ and $b$. We must estimate them from actual market data.

This part introduces the statistical tool most commonly used for this purpose: ***regression analysis***. We will learn how to:

- Estimate demand equations from data

- Interpret the estimated coefficients

- Assess how well our model fits the data

- Use the model for prediction

(b) **What is Regression?**

Regression analysis is a statistical method that finds the mathematical relationship between a dependent variable (***what we want to explain***) and one or more independent variables (***the factors that explain it***).

For demand, the dependent variable is quantity demanded ($Q$) , and the independent variable is price ($P$) . Regression finds the straight line (or curve) that best fits the observed data points.

(c) **The Simple Linear Regression Model**
The simple linear regression model is written as:

$$
Q_i = \alpha + \beta P_i + \epsilon_i
$$


$Q_i$ is the quantity demanded for observation $i$

$P_i$ is the price for observation $i$

$\alpha$ is the true (unknown) intercept

$\beta$ is the true (unknown) slope

$\varepsilon_i$ is the error term – the difference between the actual $Q_i$ and the value predicted by the line $\alpha + \beta P_i$

The error term captures everything else that affects quantity but is not included in the model (e.g., changes in income, tastes, weather, random chance). A key assumption is that these errors are random and average out to zero.

(d) **The Goal of Regression**

The goal is to use the data to find estimates of $\alpha$ and $\beta$, which we denote as $\hat{\alpha}$ and $\hat{\beta}$. These estimates define the estimated regression line:

$$
\hat{Q}_i = \hat{\alpha} + \hat{\beta} P_i
$$

where $\hat{Q}_i$ is the predicted quantity for observation $i$.

(e) **Ordinary Least Squares (OLS)**

The most common method for estimating $\alpha$ and $\beta$ is Ordinary Least Squares (OLS) . OLS chooses the estimates that minimize the sum of squared errors (the differences between actual and predicted values).

(f) **The Idea of Least Squares**

For each data point, the residual $e_i$ is the difference between the actual quantity $Q_i$ and the predicted quantity $\hat{Q}_i$:

$$
e_i = Q_i - \hat{Q}_i = Q_i - (\hat{\alpha} + \hat{\beta}P_i)
$$


OLS finds $\hat{\alpha}$ and $\hat{\beta}$ that minimize the sum of squared residuals (SSR) :

$$
SSR = \sum_{i=1}^{n} e_i^2 = \sum_{i=1}^{n} (Q_i - \hat{\alpha} - \hat{\beta}P_i)^2
$$

Why square the residuals? Squaring ensures that positive and negative errors do not cancel each other out and places a heavier penalty on large errors.

(g) **Formulas for OLS Estimates**

Without diving into calculus, the OLS estimates are given by:

$$
\hat{\beta} = \frac{\sum_{i=1}^{n}(P_i - \bar{P})(Q_i - \bar{Q})}{\sum_{i=1}^{n}(P_i - \bar{P})^2}
$$

$$
\hat{\alpha} = \bar{Q} - \hat{\beta}\bar{P}
$$

where $\bar{P}$ and $\bar{Q}$ are the sample means of price and quantity.

***In practice***, we never compute these by hand; we use statistical software. But the formulas remind us that the slope estimate is essentially the covariance between $P$ and $Q$ divided by the variance of $P$.

(h) **Estimating a Linear Demand Model**

A Step‑by‑Step Example
Let us return to the cassava data:

| Price (P) | Quantity (Q) |
| :--- | :--- |
| 1 | 100 |
| 2 | 70 |
| 4 | 50 |
| 8 | 35 |

We will estimate a linear demand model $Q = \alpha + \beta P$ using Python.

```python
# Cassava data
data = pd.DataFrame({
    'P': [1, 2, 4, 8],
    'Q': [100, 70, 50, 35]
})

# Add a constant term for the intercept
X = sm.add_constant(data['P'])  # X now has two columns: const and P
y = data['Q']

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())
```

**The fitted equation**

```python
# the fitted equation clearly
print("Intercept:", model.params['const'])
print("Slope:", model.params['P'])
print(f"Estimated equation: Q = {model.params['const']:.2f} + ({model.params['P']:.2f})P")
```

\hat{Q} = 94.565 - 8.217P

***Slope (-8.217):*** For each 1‑cedi increase in price, quantity demanded falls by approximately ***8.22 units***.

***Goodness of fit:*** ***$R^2 = 0.820$***, meaning that price explains ***82%*** of the variation in quantity. The remaining 18% is due to other factors.

This is the estimated value of $\ln Q$ when $\ln P = 0$ (i.e., when ***$P = 1$***). Exponentiating gives $\hat{Q} = e^{4.6052} \approx 100.0$, which matches the actual data point perfectly.

***Slope ($\hat{b}$)*** = -0.5054
This is the estimated price elasticity of demand. A **1%** increase in price leads to an approximate ***0.505%*** decrease in quantity demanded. Since the absolute value is less than 1, cassava demand is inelastic – consumers are not very responsive to price changes.

***Statistical Significance:*** The t‑statistic for lnP is ***-45.12***, with a p‑value essentially zero. The 95% confidence interval for elasticity is ***$[-0.554, -0.457]$***.

***Goodness of Fit:*** $R^2$ = 1.000$ (rounded to three decimal places). This means that ***100%*** of the variation in $\ln Q$ is explained by $\ln P$ – a perfect fit. The residuals are essentially zero, indicating that the data exactly follow the constant‑elasticity form.

- **Linear Model – Fit and Residuals**

***What it shows:***

**Left panel:** The actual cassava data points (Price on vertical axis, Quantity on horizontal axis) and the fitted linear demand line $\hat{Q} = 94.565 - 8.217P$.

**Right panel:** Residuals (actual quantity minus predicted quantity) plotted against price.

***Key insights:***

The straight line does not pass through any of the data points; it is the best‑fitting line according to ordinary least squares, but the fit is only moderate ($R^2 = 0.820$).

Residuals show that the model over‑predicts at some prices and under‑predicts at others. The residuals are relatively large (up to ±13.6 units) and alternate in sign.

The pattern of residuals suggests that a straight line may not be the ideal functional form for this data.



```python
# Linear model coefficients
intercept_lin = 94.565
slope_lin = -8.217

# Predicted values and residuals
data['Q_pred_lin'] = intercept_lin + slope_lin * data['P']
data['resid_lin'] = data['Q'] - data['Q_pred_lin']

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Left: data and fitted line
ax1.scatter(data['P'], data['Q'], s=100, label='Actual data')
P_grid = np.linspace(0, 9, 100)
Q_grid = intercept_lin + slope_lin * P_grid
ax1.plot(P_grid, Q_grid, linewidth=2,
         label='Linear fit: Q = {:.2f} + ({:.2f})P'.format(intercept_lin, slope_lin))
ax1.set_xlabel('Price (P)')
ax1.set_ylabel('Quantity (Q)')
ax1.set_title('Linear Model: Fit to Data')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Right: residual plot
ax2.scatter(data['P'], data['resid_lin'], s=100)
ax2.axhline(y=0, linestyle='--', linewidth=1)
ax2.set_xlabel('Price (P)')
ax2.set_ylabel('Residuals (Q - Q̂)')
ax2.set_title('Residual Plot – Linear Model')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

 - **Log‑Log Model – Transformed and Original Scales**

***What it shows:***

**Left panel:** The transformed data ($\ln P$, $\ln Q$) with the fitted line $\ln\hat{Q} = 4.6031 - 0.5029\ln P$.

**Right panel:** The original data with the fitted log‑log curve back‑transformed to $Q = e^{4.6031} P^{-0.5029}$.

***Key insights:***

In the log‑log space (left), the points fall exactly on a straight line, explaining the $R^2 = 1.000$ (perfect fit).

When transformed back to the original scale (right), the curve passes through every data point, confirming that the data follow a constant‑elasticity (multiplicative) pattern perfectly.

This graph demonstrates why the log‑log model is preferred when the relationship is multiplicative rather than additive.



```python
# Log-log coefficients
intercept_ll = 4.6031
slope_ll = -0.5029

# Predicted values
data['lnQ_pred'] = intercept_ll + slope_ll * data['lnP']
data['Q_pred_ll'] = np.exp(data['lnQ_pred'])
data['resid_ll'] = data['lnQ'] - data['lnQ_pred']   # residuals on log scale

# Create figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Left: log-log space
ax1.scatter(data['lnP'], data['lnQ'], s=100, label='Transformed data')
lnP_grid = np.linspace(-0.2, 2.2, 100)
lnQ_grid = intercept_ll + slope_ll * lnP_grid
ax1.plot(lnP_grid, lnQ_grid, linewidth=2,
         label='ln Q = {:.4f} + ({:.4f}) ln P'.format(intercept_ll, slope_ll))
ax1.set_xlabel('ln(Price)')
ax1.set_ylabel('ln(Quantity)')
ax1.set_title('Log-Log Model: Linear on Log Scale')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Right: back-transformed to original scale
ax2.scatter(data['Q'], data['P'], s=100, label='Actual data')
P_smooth = np.linspace(0.5, 9, 200)
Q_smooth = np.exp(intercept_ll) * (P_smooth ** slope_ll)
ax2.plot(Q_smooth, P_smooth, linewidth=2,
         label='Q = {:.2f} × P^({:.4f})'.format(np.exp(intercept_ll), slope_ll))
ax2.set_xlabel('Quantity (Q)')
ax2.set_ylabel('Price (P)')
ax2.set_title('Log-Log Model: Original Scale')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_xlim(0, 110)

plt.tight_layout()
plt.show()
```
